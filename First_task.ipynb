```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
date_cols = ['date']
```

```python
data = pd.read_csv('./Data/ccc_compiled_2017-2020_fipscompiled.csv', encoding = 'latin-1', low_memory=False, parse_dates=date_cols)
```

```python
data
```

```python
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
```

```python
data.loc[data['month'] <= 3, 'quantile'] = int(1)
data.loc[(data['month'] >= 4) & (data['month'] <= 6), 'quantile'] = int(2)
data.loc[(data['month'] >= 7) & (data['month'] <= 9), 'quantile'] = int(3)
data.loc[(data['month'] >= 10) & (data['month'] <= 12), 'quantile'] = int(4)
```

```python
data
```

```python
root_dir = 'Data/ZIPCBSA/'
zip_cbsa_files = os.listdir(root_dir)
```

```python
dataframes = []

for file in tqdm(zip_cbsa_files, total = len(zip_cbsa_files)):
    file_path = os.path.join(root_dir, file)
    df = pd.read_excel(file_path)
    df.columns = df.columns.str.lower()

    file_name, extension = os.path.splitext(os.path.basename(file))
    df['source'] = file_name

    dataframes.append(df)
```

```python
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
combined_df
```

```python
combined_df['year'] = combined_df['source'].apply(lambda x : x[:4])
combined_df['quantile'] = combined_df['source'].apply(lambda x : x[5:])
```

```python
combined_df.drop(columns = ['source'], inplace = True)
```

```python
combined_df
```

```python
# if it's from 2017 and Jan 1 ~ Mar 31, then go for 2017q1.xlsx file. and then get zip code which corresponds with cbsa
```

```python
# Convert columns in both dataframes to the same type (e.g., str)
data['year'] = data['year'].astype(int)
data['quantile'] = data['quantile'].astype(int)
data['cbsa'] = data['cbsa'].astype(float)

combined_df['year'] = combined_df['year'].astype(int)
combined_df['quantile'] = combined_df['quantile'].astype(int)
combined_df['cbsa'] = combined_df['cbsa'].astype(float)
```

```python
new_df = pd.merge(data, combined_df[['year', 'quantile', 'cbsa', 'zip']], 
                how='left', 
                on=['year', 'quantile', 'cbsa'])
```

```python
# Perform the left merge to bring in the 'zip' column from combined_df
data = pd.merge(data, combined_df[['year', 'quantile', 'cbsa', 'zip']], 
                how='left', 
                on=['year', 'quantile', 'cbsa'])

# Rename the 'zip' column to 'zipcode'
data.rename(columns={'zip': 'zipcode'}, inplace=True)
```
