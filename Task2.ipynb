## Small task for taking unique issue (Task 2)

```python
date_cols = ['date']
```

```python
data = pd.read_csv('./Data/ccc_compiled_2017-2020_fipscompiled.csv', encoding = 'latin-1', low_memory=False, parse_dates=date_cols)
```

```python
# Split the 'issues' column by ';' and expand into multiple columns
split_issues = data['issues'].str.split(';', expand=True)
```

```python
# Remove any extra whitespace from split values
split_issues = split_issues.apply(lambda x: x.str.strip())
```

```python
# Rename columns dynamically (issues1, issues2, issues3, ...)
split_issues.columns = [f'issues{i+1}' for i in range(split_issues.shape[1])]
```

```python
split_issues[split_issues['issues15'].notna()]
```

```python
data.iloc[69305]
```

```python
# Flatten the DataFrame into a single list and remove NaN values
all_issues = pd.Series(split_issues.values.ravel()).dropna()
```

```python
# Get unique values from the list
unique_issues = all_issues.unique()
```

```python
# Create a new DataFrame with one column for the unique values
unique_issues_df = pd.DataFrame(unique_issues, columns=['unique_issues'])

# Display the new DataFrame with unique values
print(unique_issues_df)
```

```python
unique_issues_df.to_csv('Output/unique issues.csv', index = False)
```

```python
# Concatenate the new columns with the original DataFrame
data = pd.concat([data, split_issues], axis=1)
```

```python
data.to_csv('Output/ccc_data_including_issues.csv', index = False)
```
